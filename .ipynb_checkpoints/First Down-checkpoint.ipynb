```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.mode.chained_assignment = None
```

```python
# read in data
data = pd.DataFrame()

for year in range(1999,2021):  
    i_data = pd.read_csv('https://github.com/guga31bb/nflfastR-data/blob/master/data/' \
                         'play_by_play_' + str(year) + '.csv.gz?raw=True',
                         compression='gzip', low_memory=False)

    data = data.append(i_data)
```

```python
# select only first down and only regular season plays
df = data.loc[(data['down']==1)&(data['season_type']=='REG')]

# select only neutral game script plays (score differential between 0-7)
df_neutral = df[(df['score_differential']<=7) & (df['score_differential']>=-7)]
```

```python
### BASIC STATISTICS

# grouping by year
df_neutral_year = df_neutral.groupby('season')[['rush_attempt','pass_attempt','complete_pass','interception']].sum()


# adding completion percentage
df_neutral_year['completion_pctg'] = (df_neutral_year['complete_pass']/df_neutral_year['pass_attempt']) * 100
df_neutral_year['completion_pctg'] = df_neutral_year['completion_pctg'].round(2)

# adding passing frequency
df_neutral_year['pass_freq'] = (df_neutral_year['pass_attempt']/(df_neutral_year['pass_attempt']
                                                              +df_neutral_year['rush_attempt']))*100
df_neutral_year['pass_freq'] = df_neutral_year['pass_freq'].round(2)

# interception rate
df_neutral_year['int_rate'] = (df_neutral_year['interception']/df_neutral_year['pass_attempt'])*100
df_neutral_year['int_rate'] = df_neutral_year['int_rate'].round(2)

df_neutral_year = df_neutral_year[['completion_pctg','int_rate','pass_freq']]

df_neutral_year
```

```python
# increase percentage for completion rate, interception rate and pass frequency
for column in df_neutral_year.columns.tolist():
    final = df_neutral_year[column].iloc[-1]
    initial = df_neutral_year[column].iloc[0]

    increase = round((final-initial)/initial*100,2)

    print(str(column)+':',str(increase)+'%')
```

```python
# Pass Completion, Interception Rate and Pass Frequency #single plot
plt.figure(figsize = (18,8))

for column in df_neutral_year.columns.tolist():

    
    ax = df_neutral_year[column].plot()

    # set title
    ax.set_title('Pass Completion, Interception Rate and Pass Frequency over the Years', fontsize=20, pad=30)

    # set y label title
    ax.set_ylabel('Percentage', fontsize=15)

    # set x label title
    ax.set_xlabel('Season', fontsize=15)

    # remove top and right border
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)  

# set legend    
plt.legend(['Pass Completion %','Interception Rate %','Pass Frequency %'], loc=5, frameon=False, fontsize=12)



plt.show()
```

```python
# Pass Completion, Interception Rate and Pass Frequency #multiple plot
df_neutral_year_renamed = df_neutral_year.rename(columns={'completion_pctg':'Completion Percentage','int_rate':'Interception Rate',
                                'pass_freq':'Pass Frequency'})

for column in df_neutral_year_renamed.columns.tolist():

    plt.figure(figsize = (9,6))
    ax = df_neutral_year_renamed[column].plot(c='dodgerblue')

    # set title
    ax.set_title('{}'.format(column), fontsize=20, pad=30)

    # set y label title
    ax.set_ylabel('Percentage', fontsize=15)

    # set x label title
    ax.set_xlabel('Season', fontsize=15)

    # remove top and right border
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)  


plt.show()
```

### Advanced Statistics

Before reading on, keep in mind:


- **YPA**: yards per attempt, that is an average of how many yards that type of play gains


- **EPA**: expected points added, that is how many points the offense is expected to score taking into consideration the distance to goal line, the distance to the first down marker and down


- **Frequency of Negative EPA**: how often that type of play produces negative EPA


- **Success Rate**: plays that gain at least 40% of the yards required for next set of down are considered successful (e.g. 4 yards on 1st & 10)

```python
# calculate average EPA for a run by year
neutral_run = df_neutral.loc[df_neutral['play_type']=='run']
neutral_run_epa = neutral_run.groupby('season')[['epa']].mean().round(3).rename(columns={'epa':'run_epa'})

# calculate average EPA for a pass by year
neutral_pass = df_neutral.loc[df_neutral['play_type']=='pass']
neutral_pass_epa = neutral_pass.groupby('season')[['epa']].mean().round(3).rename(columns={'epa':'pass_epa'})

# calculate frequency of a negative EPA run by year
neutral_run_epa_freq = neutral_run.groupby('season')['epa'].apply(lambda x: pd.Series([round((x < 0).sum()/len(x)*100,2)])).unstack()
neutral_run_epa_freq = neutral_run_epa_freq.rename(columns={0:'run_epa_freq'})

# calculate frequency of a negative EPA pass by year
neutral_pass_epa_freq = neutral_pass.groupby('season')['epa'].apply(lambda x: pd.Series([round((x < 0).sum()/len(x)*100,2)])).unstack()
neutral_pass_epa_freq = neutral_pass_epa_freq.rename(columns={0:'pass_epa_freq'})

# calculate YPA of rushes by year
neutral_run_ypa = neutral_run.groupby('season')[['yards_gained']].mean().round(2).rename(columns={'yards_gained':'run_ypa'})

# calculate YPA of passes by year
neutral_pass_ypa = neutral_pass.groupby('season')[['yards_gained']].mean().round(2).rename(columns={'yards_gained':'pass_ypa'})

# calculate Success Rate of passes by year
neutral_run_success = neutral_run.groupby('season')['success'].apply(lambda x: pd.Series([round((x == 1).sum()/len(x)*100,2)])).unstack()
neutral_run_success = neutral_run_success.rename(columns={0:'run_success'})

# calculate Success Rate of passes by year
neutral_pass_success = neutral_pass.groupby('season')['success'].apply(lambda x: pd.Series([round((x == 1).sum()/len(x)*100,2)])).unstack()
neutral_pass_success = neutral_pass_success.rename(columns={0:'pass_success'})

# concatenate the all EPA dfs
df_neutral_epa = pd.concat([neutral_run_epa, neutral_pass_epa,
                            neutral_run_epa_freq, neutral_pass_epa_freq, 
                            neutral_run_ypa, neutral_pass_ypa,
                            neutral_run_success, neutral_pass_success], axis=1)

df_neutral_epa
```

```python
# EPA over years Pass vs. Run
plt.figure(figsize = (18,8))

ax = df_neutral_epa['run_epa'].plot()
ax2 = df_neutral_epa['pass_epa'].plot()

# set title
ax.set_title('Pass and Rush EPA over the Years', fontsize=20, pad=30)

# set y label title
ax.set_ylabel('EPA', fontsize=14)

# set x label title
ax.set_xlabel('Year', fontsize=14)

# set legend
plt.legend(['Run','Pass'], loc=5, frameon=False, fontsize=12)

# remove top and right border
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
```

```python
# Frequency of Negative EPA Play Pass vs. Run
plt.figure(figsize = (18,8))

ax = df_neutral_epa['run_epa_freq'].plot()
ax2 = df_neutral_epa['pass_epa_freq'].plot()

# set title
ax.set_title('Pass and Rush Negative EPA Frequency over the Years', fontsize=20, pad=30)

# set y label title
ax.set_ylabel('Frequency', fontsize=14)

# set x label title
ax.set_xlabel('Year', fontsize=14)

# set legend
plt.legend(['Run','Pass'], loc=5, frameon=False, fontsize=12)

# remove top and right border
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
```

```python
# Yards per Attempt Pass vs. Run
plt.figure(figsize = (18,8))

ax = df_neutral_epa['run_ypa'].plot()
ax2 = df_neutral_epa['pass_ypa'].plot()

# set title
ax.set_title('Pass and Rush YPA over the Years', fontsize=20, pad=30)

# set y label title
ax.set_ylabel('Yards per Attempt', fontsize=14)

# set x label title
ax.set_xlabel('Year', fontsize=14)

# set legend
plt.legend(['Run','Pass'], loc=5, frameon=False, fontsize=12)

# remove top and right border
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
```

```python
# Success Rate Pass vs. Run
plt.figure(figsize = (18,8))

ax = df_neutral_epa['run_success'].plot()
ax2 = df_neutral_epa['pass_success'].plot()

# set title
ax.set_title('Pass and Rush Success Rate over the Years', fontsize=20, pad=30)

# set y label title
ax.set_ylabel('Success Rate', fontsize=14)

# set x label title
ax.set_xlabel('Year', fontsize=14)

# set legend
plt.legend(['Run','Pass'], loc=5, frameon=False, fontsize=12)

# remove top and right border
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.show()
```

```python
# average EPA of the last 5 years
last5 = df_neutral_epa.tail(5).mean().round(2)

# difference between pass and run
epa = round((last5['pass_epa'] - last5['run_epa'])/abs(last5['run_epa'])*100,2)
neg_epa_freq = round((last5['pass_epa_freq'] - last5['run_epa_freq'])/abs(last5['run_epa_freq'])*100,2)
ypa = round((last5['pass_ypa'] - last5['run_ypa'])/abs(last5['run_ypa'])*100,2)
success = round((last5['pass_success'] - last5['run_success'])/abs(last5['run_success'])*100,2)

print('EPA: ' + str(epa) + '%',
      'Neg.EPA: ' + str(neg_epa_freq)+ '%',
      'YPA: ' + str(ypa) + '%',
      'Success: ' + str(success) + '%', sep = '\n')
```

### Findings

On first downs, **passes** in the last 5 years:
- produce on average ***+0.190 EPA*** more than rushes (+237.5%; 0.110 vs. -0.080)


- are ***20.7% less likely*** to produce a ***negative EPA*** play than rushes (50.54% vs. 63.73%)


- produce ***+2.7 yards per attempt*** more than rushes (+61.09%; 7.12 vs. 4.42)


- are ***36.28% more*** likely to be ***successful*** than rushes (49.43% vs. 36.27%)

furthermore, in the last 20 years:


- overall ***completion percentage increased by 17%***


- overall ***interception rate</font> <font color=green>decreased by 40.23%*</font>


- **however** *pass frequency* on first downs <font color=red>*increased by only 2.81%*</font>

In short, **rushes** on average:

- leave the team in a <font color=red>worse position</font> than they were before in terms of EPA


- are <font color=red>more likely</font> to produce a <font color=red>negative EPA play</font>


- gain <font color=red>fewer yards per attempt</font>


- are <font color=red> less successful</font> in gaining 40% of the yards needed for the first down


and yet:


- are still <font color=red>*called more often than passes*</font> (53% vs. 47% in the last 5 years)
